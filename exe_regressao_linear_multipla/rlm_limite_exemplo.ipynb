```python
# Bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
# Importando a base
limite = pd.read_excel(r"C:\Users\alexa\OneDrive\Área de Trabalho\repositorios_git\DNC_repository\exe_regressao_linear_multipla\limite_credito.xlsx")
limite.head()
```

```python
# Correlação
limite.corr()
```

```python
# Análise estatisticas das colunas
sns.pairplot(limite);
```

```python
#modelo MRLS
function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato'
# primeira variável é a preditiva, separado por '~' e o restante são as explicativas, separadas por '+'
model = smf.ols(formula=function, data=limite).fit() 
print(model.summary())
```

```python
#Remover a variável RendimentoTotal que tem o maior P-value
#Ainda temos variáveis acima de 5%
function = 'LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato'
model = smf.ols(formula=function, data=limite).fit() 
print(model.summary())
```
